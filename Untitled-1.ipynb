```python
import os
import zipfile
import shutil

# Paths
outer_zip_path = "C:/Users/jimmy/Downloads/6337421.zip"
extract_dir = "C:/Users/jimmy/Desktop/Bachelor Thesis Project/dataset"
audio_target_dir = os.path.join(extract_dir, "audio")
dcase25_path = "C:/Users/jimmy/Desktop/Bachelor Thesis Project/dataset/dcase25.py"

# Files to verify
test_files = [
    "tram-lisbon-1100-43664-2-a.wav",
    "tram-prague-1071-40104-6-a.wav",
    "airport-helsinki-1-1-a.wav"
]

def unzip_nested_dataset(outer_zip_path, extract_dir, audio_target_dir):
    """Unzip the outer ZIP, inner ZIPs, and merge audio files from nested structure."""
    # Clean up existing dataset
    if os.path.exists(extract_dir):
        print(f"Removing existing folder: {extract_dir}")
        shutil.rmtree(extract_dir)
    os.makedirs(audio_target_dir, exist_ok=True)

    # Step 1: Extract the outer ZIP
    temp_outer_dir = os.path.join(extract_dir, "temp_outer")
    os.makedirs(temp_outer_dir, exist_ok=True)
    print(f"Unzipping outer ZIP {outer_zip_path} to {temp_outer_dir}...")
    with zipfile.ZipFile(outer_zip_path, 'r') as zip_ref:
        zip_ref.extractall(temp_outer_dir)

    # Step 2: Find and extract inner ZIPs
    inner_zip_files = [f for f in os.listdir(temp_outer_dir) if f.endswith(".zip")]
    if not inner_zip_files:
        raise FileNotFoundError(f"No inner ZIP files found in {temp_outer_dir}")

    print(f"Found {len(inner_zip_files)} inner ZIP files: {inner_zip_files}")

    for i, inner_zip in enumerate(inner_zip_files):
        inner_zip_path = os.path.join(temp_outer_dir, inner_zip)
        temp_inner_dir = os.path.join(extract_dir, f"temp_inner_{i}")
        os.makedirs(temp_inner_dir, exist_ok=True)
        
        print(f"Unzipping inner ZIP {inner_zip_path} to {temp_inner_dir}...")
        with zipfile.ZipFile(inner_zip_path, 'r') as zip_ref:
            zip_ref.extractall(temp_inner_dir)

        # Move WAV files from nested TAU-urban-acoustic-scenes-2022-mobile-development/audio/
        nested_audio_dir = os.path.join(temp_inner_dir, "TAU-urban-acoustic-scenes-2022-mobile-development", "audio")
        if os.path.exists(nested_audio_dir):
            print(f"Merging audio files from {nested_audio_dir}...")
            for file in os.listdir(nested_audio_dir):
                if file.endswith(".wav"):
                    src = os.path.join(nested_audio_dir, file)
                    dst = os.path.join(audio_target_dir, file)
                    if not os.path.exists(dst):
                        shutil.move(src, dst)
                    else:
                        print(f"Skipping duplicate: {file}")
        else:
            print(f"No nested audio/ folder found in {temp_inner_dir}")

    # Clean up temporary directories
    shutil.rmtree(temp_outer_dir)
    for i in range(len(inner_zip_files)):
        temp_inner_dir = os.path.join(extract_dir, f"temp_inner_{i}")
        if os.path.exists(temp_inner_dir):
            shutil.rmtree(temp_inner_dir)

def verify_files(audio_dir, test_files):
    """Verify that key audio files exist."""
    print("Verifying audio files...")
    for file in test_files:
        full_path = os.path.join(audio_dir, file)
        if os.path.exists(full_path):
            print(f"Found: {full_path}")
        else:
            print(f"Missing: {full_path}")
            return False
    return True

def update_dcase25(dcase25_path, extract_dir):
    """Update dataset_dir in dcase25.py."""
    new_dataset_dir = extract_dir.replace("/", "\\\\")
    new_line = f'dataset_dir = "{new_dataset_dir}\\"\n'

    print(f"Updating {dcase25_path} with dataset_dir = {new_dataset_dir}")
    with open(dcase25_path, 'r') as f:
        lines = f.readlines()

    updated = False
    for i, line in enumerate(lines):
        if line.strip().startswith("dataset_dir ="):
            lines[i] = new_line
            updated = True
            break
    
    if not updated:
        lines.insert(0, new_line)

    with open(dcase25_path, 'w') as f:
        f.writelines(lines)

def main():
    # Step 1: Unzip and merge nested audio files
    unzip_nested_dataset(outer_zip_path, extract_dir, audio_target_dir)
    
    # Step 2: Verify key files
    if verify_files(audio_target_dir, test_files):
        print("All test files found successfully.")
    else:
        print("Some test files are missing. Check the ZIP file or extraction process.")
        return

    # Step 3: Update dcase25.py
    update_dcase25(dcase25_path, extract_dir)
    print("Dataset setup completed.")

if __name__ == "__main__":
    if not os.path.exists(outer_zip_path):
        print(f"Error: Outer ZIP file not found at {outer_zip_path}")
    else:
        main()
```
